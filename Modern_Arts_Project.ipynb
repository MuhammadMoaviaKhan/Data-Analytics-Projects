```python
def remove_bad_char(bad_str):
  bad_char = ['c', '.', ' ', "'", 's','(', ')', 'C']
  for i in bad_char:
    bad_str = bad_str.replace(i, "")
  return bad_str
```

```python
date = '2000-2008'

if '-' in date:
  date = date.split('-')
  yr1 = int(date[0])
  yr2 = int(date[1])
  yr = (yr1+yr2)/2
  print(round(yr))
```

```python
def clean_date(date):
  if '-' in date:
    lst = date.split('-')
    yr1 = int(lst[0])
    yr2 = int(lst[1])
    date = round((yr1+yr2)/2)
  else:
    date = int(date)
  return date
```

```python
clean_date('2000-2008')
```

```python
arts['Date'] = arts['Date'].apply(clean_date)
```

# **Artist's Career Duration:**

**What is the average career duration of artists (from begin date to end date)?**


**Are there any outliers with significantly longer or shorter career durations?**

```python
data_set = arts
```

```python
data_set.head()
```

```python
data_set.info()
```

```python
data_set.dropna(inplace = True)
```

```python
data_set.info()
```

```python
data_set['Career Duration'] = data_set['EndDate'] - data_set['BeginDate']
```

```python
data_set.head()
```

```python
data_set['Career Duration'].value_counts()
```

```python
mean_duration = data_set['Career Duration'].mean()
print(round(mean_duration))
```

```python
shortest_career = 100
longest_career = 0
for i in data_set['Career Duration']:
  if longest_career < i:
    longest_career = i
  elif shortest_career > i:
    shortest_career = i
  else:
    continue

print(shortest_career)
print(longest_career)
```

```python
plt.figure(figsize = (10, 6))
sns.violinplot(x='Career Duration', data=data_set, color = 'cyan')
plt.title('\nDistribution of Artists\' Career Durations\n')
plt.xlabel('\nCareer Duration (Years)')
plt.show()
print(f'Mean of Artists Career is: {round(mean_duration, 1)}')
print(f'Longest Career of Artist acording to this data is: {longest_career} ')
print(f'Shortest Career of Artist acording to this data is: {shortest_career} ')
```

# **Comparison of Gender Representation Over Time**
**How has the gender representation in the art world evolved over time?**


**Are there any noticeable changes in the proportion of male and female artists over different periods?**

```python
arts.head(2)
```

```python
gender_ditribution = arts.groupby(['BeginDate', 'Gender']).size().reset_index(name='Count')
```

```python
plt.figure(figsize = (10, 6))
sns.lineplot(x = 'BeginDate', y = 'Count', hue = 'Gender', data = gender_ditribution, palette = 'bright')
plt.title('\nGender Representation Over Time\n', fontsize = 15)
plt.xlabel('\nYear of Starting Career')
plt.ylabel('Counts of Gender\n')
plt.show()
```

```python
plt.figure(figsize = (10, 6))
sns.violinplot(x = 'EndDate', y = 'Gender', hue = 'Gender', data = arts, palette = 'YlOrBr')
plt.title('\n Gender vs Endate\n', fontsize = 15)
plt.xlabel('\nEnd Date')
plt.ylabel('\nGenders\n')
plt.show()
```
