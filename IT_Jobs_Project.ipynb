# DATA VISUALIZATION

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['salary_in_usd'], bins = 30, kde=True)
plt.title('Salary Distribution')
plt.xlabel('Salary (USD)')
plt.ylabel('Frequency')
# plt.grid()
plt.show()
```

```python
employment_type =  df['employment_type'].value_counts()
# employment_type.values
# employment_type.index
plt.figure(figsize=(12, 6))
colour = ['darkcyan', 'darkgoldenrod', 'darkgray', 'darkgreen']
plt.barh(employment_type.index, employment_type.values, color = colour)
plt.title('Distribution of Employment Types')
plt.xlabel('Count')
plt.ylabel('Employment Type')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x= 'salary_currency', y= 'salary_in_usd', data = df)
plt.title('Distribution of Salary Currencies')
plt.xlabel('Salary Currency')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 10))
experience_level_counts = df['experience_level'].value_counts()
plt.pie(experience_level_counts, labels=experience_level_counts.index, autopct= '%1.1f%%', startangle=140)
plt.title('Distribution of Experience Levels (Pie Chart)')
plt.show()

# % is a literal percentage sign.
# 1.1f specifies the format for the percentage value
# 1 before the dot specifies the minimum width of the field.
# . is a decimal point.
# 1 after the dot specifies the number of digits after the decimal point.
```

```python
top_10_job_title_according_salaries = df.sort_values(by= ['salary_in_usd'], ascending = False)
ds = top_10_job_title_according_salaries.head(10)

plt.figure(figsize=(30, 15))
colour = ['darkcyan', 'darkgoldenrod', 'darkgray', 'darkgreen']
sns.scatterplot(x= 'job_title', y= 'salary_in_usd', data= ds, s= 300)
plt.title('(TOP 10) Distribution of Salary According To Job Title ')
plt.xlabel('Job Title')
plt.ylabel('Salary in USD')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x= 'work_year', y= 'salary_in_usd', data = df, err_style = None, color= 'darkcyan', marker = 'o')
plt.title('Salary Trends Over the Years')
plt.xlabel('Work Year')
plt.ylabel('Salary in (USD)')
plt.grid()
plt.show()
```
