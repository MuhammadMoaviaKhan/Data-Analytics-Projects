***Selling Percentage***

```python
most_selling_brands = cars['brand'].value_counts(normalize = True)
most_selling_brands_bol = most_selling_brands > 0.01
```

```python
brand_bol = most_selling_brands[most_selling_brands_bol]
brand_bol
```

```python
top_bol = cars['brand'].isin(brand_bol.index)
```

***Top 10 Brands by Selling Percentage***

```python
top_selling_brand = cars[top_bol].copy()
```

```python
top_selling_brand.head()
```

***TypeCast and Clean the column of (Price & Odometer)***

```python
top_selling_brand['price'] = top_selling_brand['price'].str.replace('$', '').str.replace(',', '').astype(int)
top_selling_brand['odometer'] = top_selling_brand['odometer'].str.replace('km', '').str.replace(',', '').astype(int)
```

```python
top_selling_brand.head(3)
```

```python
print(top_selling_brand['price'].dtype)
print(top_selling_brand['odometer'].dtype)
```

```python
top_selling_brand.rename({
    'price' : 'price ($)',
    'odometer' : 'odometer (km)'
}, axis = 1, inplace = True)
```

```python
top_selling_brand.columns
```

```python
grouping = top_selling_brand.groupby(['brand'])
```

```python
mean_price = grouping['price ($)', 'odometer (km)'].mean()
```

```python
mean_price.sort_values('price ($)', ascending = False)
```

```python
top_selling_brand['price ($)'].value_counts().sort_index()
```

```python
clean_price = top_selling_brand['price ($)'].between(1, 501000)
```

```python
top_selling_brand = top_selling_brand[clean_price]
```

```python
top_selling_brand.head()
```

```python
top_selling_brand['yearOfRegistration'].value_counts().sort_index()
```

```python
clean_year_of_registration = top_selling_brand['yearOfRegistration'].between(1900, 2016)
```

```python
top_selling_brand = top_selling_brand[clean_year_of_registration]
```

```python
print(top_selling_brand['yearOfRegistration'].max())
print(top_selling_brand['yearOfRegistration'].min())
```
