## DATA VISUALIZATION

```python
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
px.pie(top, names= 'Player', values= 'Runs', title= 'Highest Runs Holders', hover_data='Ave')
```

```python
px.bar(top, x= 'Player', y= 'Ave', height=400, color= 'Player', title= 'Best Batting Average')
```

```python
most_sixes = df.sort_values(by= ['6s'], ascending= False)
sixes= most_sixes.head(5)
sixes
```

```python
px.bar(sixes, x= 'Player', y= '6s', height=400, color= 'Player', title= 'Most Sixes')
```

```python
df['HS'] = pd.to_numeric(df['HS'].replace('[\*,]', '', regex=True), errors='coerce')
highest_score = df.sort_values(by= ['HS'], ascending= False)
highest_scores= highest_score.head(5)
highest_scores
```

```python
px.bar(highest_scores, x= 'Player', y= 'HS', height=400, color= 'Player', title= 'Highest Score In an Inning')
```

```python
most_100 = df.sort_values(by= ['100'], ascending= False)
most_100s= most_100.head(5)
most_100s
```

```python
px.scatter(most_100s, x= 'Player', y= '100', height=400, color= 'Player', title= 'Most 100 In an Inning', width= 1000)
```

```python
highest_econ= df1.sort_values(by= ['Econ'], ascending=False)
highest_economy= highest_econ.head(5)
highest_economy
```

```python
 px.line(highest_economy, x= 'Player', y= 'Econ',title= 'Highest Economy of Bowlers In a Match')
```

```python
most_wickets= df1.sort_values(by= ['Wkts'], ascending=False)
most_wicket= most_wickets.head(5)
most_wicket
```

```python
sns.barplot(data=  most_wicket, x= 'Player', y= 'Wkts')
plt.title('Most 3 Wickets Taking Bowlers In a Match')
plt.ylabel('Wickets')
plt.show()
```

```python
top_wickets_takers= df2.sort_values(by= ['Wkts'], ascending=False)
top_wickets_taker= top_wickets_takers.head(5)
top_wickets_taker
```

```python
px.pie(top_wickets_taker, names= 'Player', values= 'Wkts', title= 'Top 5 Wicket Taker Bowler In the World-cup 2023', hover_data='Wkts')
```
